```python
import pandas as pd 

df = pd.read_csv('Scorecard.txt', sep='\t',)
df.columns
```

```python
contingy_table = pd.crosstab(index=df['LOCALE 2'], columns=df['TYPE'])
contingy_table.plot(kind='bar')
```

```python
df1 = df[df['PRED.BLACK'] == 'Yes']
df2 = df[df['PRED.BLACK'] == 'No']
```

```python
contingy_table = pd.crosstab(index=df1['LOCALE 2'], columns=df1['TYPE'])
contingy_table.plot(kind='bar')
```

```python
contingy_table = pd.crosstab(index=df2['LOCALE 2'], columns=df2['TYPE'])
contingy_table.plot(kind='bar')
```
