```python
# Step by step exercie
df = pd.read_csv('./data/Housing_prices.txt', usecols=['Price', 'Age', 'Living.Area', 'Bedrooms'], sep='\t')

y = df.Price
x = df.drop('Price', axis=1)
x.Age = np.log10(x.Age+1)
```

```python
for column in list(['Living.Area', 'Age']):
    plt.title(f'Price against {column}.')
    plt.scatter(x[column], y)
    plt.show()
```

```python
for column in list(['Bedrooms']):
    prices = []
    for number in df[column].unique():
        price = df.Price[df[column] == number]
        prices.append(price)
    plt.title(f'Price against {column}.')
    plt.boxplot(prices)
    plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(x, y)
b1, b2, b3 = model.coef_
b0 = model.intercept_
predictions = model.predict(x)
```
